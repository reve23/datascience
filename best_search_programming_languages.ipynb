```python
#import libraries
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
#creating a pytrends object. request data from gooogle trends
pytrends = TrendReq(hl='en-US')

#extract data about keywords
keywords = ['Python','R','c++', 'java', 'Javascript']

pytrends.build_payload(keywords,timeframe='today 5-y')#default value

#specify the data
data = pytrends.interest_over_time()

#print out the data

#a problem when we plot the data we got error for the row of isPartial. So we need to drop that
data.drop(['isPartial'], axis=1, inplace=True)
#ploting the data
plt.plot(data)
#add a legend to understand which color represent for what
plt.legend(keywords,loc='upper left')
#we give it a name ,xaxis name and yaxis name
plt.suptitle("Programming language searches on google")
plt.xlabel('years')
plt.ylabel('weekly searches')
#lastly to show the data
plt.show()
```
